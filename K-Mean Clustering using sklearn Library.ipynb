```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
Height=np.array([185,170,168,179,182,188,180,180,183,180,180,177])
Weight=np.array([72,56,60,68,72,77,71,70,84,88,67,76])
Height=Height.reshape(-1,1)
Weight=Weight.reshape(-1,1)
```

```python
Height
```

```python
Weight
```

```python
plt.scatter(Weight,Height,marker='o',color='black',label='True')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend()
plt.show()
```

```python
X = np.hstack((Height, Weight))
```

```python
kmeans = KMeans(n_clusters=3, random_state=0, n_init=10)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('K-Means Clustering of Height-Weight Data')
plt.legend()
plt.show()
```

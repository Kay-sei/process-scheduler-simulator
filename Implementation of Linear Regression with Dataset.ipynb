```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("Housing.csv")
```

```python
df
```

```python
df.head(5)
```

```python
df.tail(5)
```

```python
X=df['area']
```

```python
X
```

```python
Y=df['price']
```

```python
Y
```

```python
X=np.array(X).reshape(-1,1)
Y=np.array(Y).reshape(-1,1)
```

```python
X
```

```python
Y
```

```python
plt.scatter(X,Y,marker="o",c='r',label="Table")
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
#Linear Regression Model
model_LR=LinearRegression()
```

```python
model_LR.fit(X,Y)
```

```python
model_LR.predict([[2500]])
```

```python
Y_pred=model_LR.predict(X)
```

```python
Y_pred
```

```python
len(Y_pred)
```

```python
plt.scatter(X,Y,c='b',label='True')
plt.plot(X,Y_pred,c='r',label='predicted')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```
